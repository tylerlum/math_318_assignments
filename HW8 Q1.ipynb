```python
import numpy as np
from tqdm import tqdm
```

```python
p = 0.5
N = 1000000
```

```python
def run_test():
    nums = [0]
    for _ in range(N):
        if np.random.uniform(low=0, high=1) < p:
            nums.append(nums[-1] + 1)
        else:
            nums.append(nums[-1] - 1)
    freq = {}
    for num in nums:
        if num in freq:
            freq[num] += 1
        else:
            freq[num] = 1
    for x in freq:
        freq[x] /= len(nums)
    return freq
```

```python
runs = 10000
counter = 0
for _ in tqdm(range(runs)):
    if run_test() > 1:
        counter += 1
print(f'Fraction return to origin = {counter / runs}')
```

```python
print(f'Expected fraction return to origin = {2*(1-p)}')
```
