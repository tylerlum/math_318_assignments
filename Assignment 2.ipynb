```python
import matplotlib.pyplot as plt
import numpy as np
```

## Question 6.a

```python
# Create geometric random variables
p = 0.01
num_geometric_random_variables = 100000

geometric_random_variables = {}
for _ in range(num_geometric_random_variables):
    i = 1
    while True:
        if np.random.random() < p:
            if i not in geometric_random_variables:
                geometric_random_variables[i] = 1
            else:
                geometric_random_variables[i] += 1
            break
        i += 1
```

```python
# Create histogram
plt.bar(geometric_random_variables.keys(), geometric_random_variables.values())
plt.xlabel('Geometric Random Variable')
plt.ylabel('Frequency')
plt.title('Geometric Random Variable Histogram')
plt.grid()
```

## Question 6.b

Briefly describe how this plot compares to the histogram from part a):

* The geometric random variable probability mass density seems to have the same general shape as the geometric random variable histogram, which is what we expected. The pmf represents the probability of different outcomes, and the theoretical probability matches the empirical results particularly well because we used a large number of random variables.

```python
# Create probability density function
values = np.linspace(start=1, stop=1000, num=1000)
pmf = (1-p)**(values-1)*p

plt.plot(values, pmf)
plt.xlabel('Geometric Random Variable')
plt.ylabel('Probability')
plt.title('Geometric Random Variable Probability Mass Density')
plt.grid()
```

Briefly compare this plot to the two plots above.

* The decaying exponential graph has the same shape as the plots in part a) and b). The only difference is that the scale of the graphs is quite different. The max value for a) is nearly 1000, for b) is 0.01, and for c) is 1.0. 

```python
# Create plot of f(t) = e^{-t}
min_t, max_t = 0, 10
num_points = 10000
t = np.linspace(start=min_t, stop=max_t, num=num_points)
f = np.exp(-t)
plt.plot(t, f)
plt.xlabel('t')
plt.ylabel('exp(-t)')
plt.title('exp(-t)')
plt.grid()
```
