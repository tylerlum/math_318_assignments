```python
import numpy as np
import matplotlib.pyplot as plt
```

# Question  6b

```python
# Setup parameters
num_periods = 10000
num_decades_per_period = 100
earthquake_parameter = 1.0

# Generate random variables
lambda_ = earthquake_parameter * num_decades_per_period
X = np.random.poisson(lam=lambda_, size=(num_periods))
```

```python
# Plot histogram
plt.hist(X, bins=100)
plt.xlabel("Number of Earthqukes in 100 decade period")
plt.ylabel("Frequency")
plt.title("Histogram of Number of Earthquakes in 100 decade period")
```

# Question 6c

```python
# Calculate M
M = np.zeros_like(X)
for i in range(len(X)):
    M[i] = np.sum(np.where(X[:i] == 100, 1, 0))
```

```python
# Plot M
plt.plot(np.arange(len(M)), M / (np.arange(len(M)) + 1))
plt.xlabel('Number of periods (i)')
plt.ylabel('Average number of exactly 100 earthquakes in period (M[i] / i)')
plt.title('(M[i] / i) vs i')
plt.grid()
```
