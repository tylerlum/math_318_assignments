```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Setup test parameters
N_steps = 20000
N_coins = 1000
```

```python
# Keep track of heads/tails of each coin
position_to_heads = [False for _ in range(N_coins)]
```

```python
# Update state over time
step_to_state = []
step_to_state.append(sum(position_to_heads))
for step in range(N_steps):
    # Flip a coin at random position
    random_position = np.random.randint(low=0, high=len(position_to_heads))
    previous_side = position_to_heads[random_position]
    new_side = np.random.choice([False, True])
    position_to_heads[random_position] = new_side
    
    # Store state
    previous_state = step_to_state[-1]
    new_state = previous_state + (int(new_side) - int(previous_side))
    step_to_state.append(new_state)
```

```python
# Histogram for t = [0, 1000]
state_0_1000 = step_to_state[:1000]
plt.hist(state_0_1000, bins=20)
plt.xlabel('Number of heads')
plt.ylabel('Frequency')
plt.title('Histogram for t in [0, 1000]')
```

```python
# Histogram for t = [1000, 2000]
state_1000_2000 = step_to_state[1000:2000]
plt.hist(state_1000_2000, bins=20)
plt.xlabel('Number of heads')
plt.ylabel('Frequency')
plt.title('Histogram for t in [1000, 2000]')
```

```python
# Histogram for t = [10000, 20000]
state_10000_20000 = step_to_state[10000:20000]
plt.hist(state_10000_20000, bins=20)
plt.xlabel('Number of heads')
plt.ylabel('Frequency')
plt.title('Histogram for t in [10000, 20000]')
```
