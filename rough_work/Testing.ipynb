```python
from tqdm import tqdm
```

```python
dice_outputs = [1, 2, 3, 4, 5, 6]
N = 1000000
groups = []
for _ in tqdm(range(N)):
    group = []
    for _ in range(5):
        group.append(np.random.choice(dice_outputs))
    groups.append(group)
```

```python
groups
```

```python
counter = 0
for group in groups:
    group_set = set(group)
    if len(group_set) == len(group):
        counter += 1
        
print(f"{counter/N}")
```

```python
counter = 0
for group in groups[:N//2]:
    group_set = set(group)
    if len(group_set) == 2:
        counter += 1
        
print(f"{counter/(N//2)}")
```

```python
counter = 0
for group in groups:
    group_set = set(group)
    if len(group_set) == 2:
        counter += 1
        
print(f"{counter/(N)}")
```
